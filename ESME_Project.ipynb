## Clustering using 'meanfun' and 'label'

```python
from sklearn.cluster import KMeans

def visualize_clusters(df, num_clusters):
    colors = ['b', 'g', 'r', 'c', 'm', 'y', 'k']
    
    for n in range(num_clusters):
        clustered_df = df[df['cluster'] == n]
        plt.scatter(clustered_df['meanfun'], clustered_df['label'], c=colors[n-1])
    plt.xlabel('meanfun', fontsize=13)
    plt.ylabel('label', fontsize=13)
    
    
kmeans = KMeans(n_clusters=2)
kmeans.fit(normalized_df[['meanfun']])
voice['cluster']=kmeans.labels_
#plt.scatter(voice_normalized, kmeans.labels_, c=kmeans.labels_, s=50, cmap='viridis')
visualize_clusters(normalized_df, 2)
#plt.scatter(normalized_df['IQR'],normalized_df['label'])
```
